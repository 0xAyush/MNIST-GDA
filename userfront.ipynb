```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
denoiser = tf.keras.models.load_model('denoiser.h5')
```

```python
autoencoder = tf.keras.models.load_model('utils/autoencoder.h5')
```

```python
mnist = tf.keras.datasets.mnist 

(x_train,y_train), (x_test,y_test) = mnist.load_data()

x_train = tf.keras.utils.normalize(x_train, axis = 1)
x_test = tf.keras.utils.normalize(x_test, axis = 1)
```

```python
from noisify import noisify
```

```python
noisy_input = noisify(x_train,28,28,0.75)
```

```python
autoencoder_output = autoencoder.predict(x_train)
denoiser_output = denoiser.predict(x_train)
```

```python
def showAEoutput(n):
    print("input")
    plt.imshow(x_train[n],cmap='gray')
    plt.show()
    print("output")
    plt.imshow(autoencoder_output[n],cmap='gray')
    plt.show()

def denoiserOutput(n):
    print("origional")
    plt.imshow(x_train[n],cmap='gray')
    plt.show()
    print("input")
    plt.imshow(noisy_input[n],cmap='gray')
    plt.show()
    print("output")
    plt.imshow(autoencoder_output[n],cmap='gray')
    plt.show()
```

```python
showAEoutput(18) # 0 18
```

```python
denoiserOutput(220) #110 #130 #220
```

```python
denoiserOutput(18)
```
