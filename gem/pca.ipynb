```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from keras import models
```

```python
import pickle
```

```python
encoder = models.load_model('encoder.h5')
decoder = models.load_model('decoder.h5')
with open('codes.dat','rb') as pf:
    code = pickle.load(pf)
```

```python
from keras.datasets import mnist

((X_train,_),_) = mnist.load_data()
```

```python
from sklearn.decomposition import PCA
```

```python
j = 32
```

```python
pca = PCA(n_components=j)
pca.fit(code)
```

```python
pca_code = pca.transform(code)
```

```python
t_code = pca.inverse_transform(pca_code)
```

```python
predictions = decoder.predict([t_code])
```

```python
print(pca)
```

```python
import math
```

```python
for i in range(10):
    n = math.floor(random.random()*len(code))
    plt.imshow(X_train[n],cmap='gray')
    plt.show()
    plt.imshow(predictions[n],cmap='gray')
    plt.show()
```

```python
import pickle
with open('pca_code.dat','wb') as pf:
    pickle.dump(pca_code,pf)

with open('pca.model.dat','wb') as pf:
    pickle.dump(pca,pf)
```
