```python
import matplotlib.pyplot as plt
import tensorflow as tf
from noisify import noisify
```

```python
mnist = tf.keras.datasets.mnist 

(x_train,y_train), (x_test,y_test) = mnist.load_data()

x_train = tf.keras.utils.normalize(x_train, axis = 1)
x_test = tf.keras.utils.normalize(x_test, axis = 1)
```

```python
n_train = noisify(x_train,28,28,0.75)
n_test = noisify(x_test,28,28,0.75)
```

```python
model = tf.keras.models.load_model('denoiser.h5')
```

```python
decoded = model.predict(x_test)
```

```python
t = 0
for x in decoded:
    plt.imshow(x_test[t],cmap='gray')
    plt.show()
    plt.imshow(n_test[t],cmap='gray')
    plt.show()
    plt.imshow(x,cmap='gray')
    plt.show()
    if(t > 12):
        break
    t += 1
```
